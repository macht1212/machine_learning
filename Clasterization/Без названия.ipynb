```python
import matplotlib.pyplot as plt
import cv2
import skimage
def draw_picture(path, title, bgr=False):
    img = cv2.imread(path)
    b, g, r = cv2.split(img) # по умолчанию cv2 почему-то отдает цвета в порядке BGR вместо RGB
    new_image = cv2.merge([r, g, b])
    plt.figure(figsize=(7, 5))
    plt.axis('off')
    plt.imshow(new_image)
    plt.title(title)
    plt.show()
    
    return b, g, r
```

```python
b, g, r = draw_picture('Алгоритмы кластеризации.jpg', 'Картинка')
```

```python
def get_clast(b, g, r, n):
    kmean = KMeans(n_clusters=n)
    kmean.fit(b)
    b = kmean.labels_
    kmean.fit(g)
    g = kmean.labels_
    kmean.fit(r)
    r = kmean.labels_
    
    return b, g, r
```

```python
b, g, r = get_clast(b, g, r, 2)
```
